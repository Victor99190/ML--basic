```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_train = np.array([460, 232, 178])
w=np.array([0,0,0,0])
b = 0
itera=4000
alfa=0.0000001
cost=cost_function(x_train,y_train,w,b)
w_d,b_d= grad_calc(x_train,y_train,w,b)
costh,w,b=calculate_grad(x_train,y_train,w,b,cost,alfa,itera)
w_2d=w.reshape(-1,4)
first_column=w_2d[:,0]
print(first_column.shape,len(costh))
plot(first_column,costh)
```

```python
def plot(w,costh):
    plt.figure(figsize=(8, 6))  # Create a new figure with a specified size
    plt.plot(w,costh, 'bo-', label='Data points')  # 'bo-' for blue color, circle markers, and solid line
    plt.title('Plot of w vs. cost')  # Title of the plot
    plt.xlabel('w')               # X-axis label
    plt.ylabel('cost')               # Y-axis label
    plt.legend()                  # Display legend
    plt.grid(True)
    plt.show()
```

```python
def cost_function(x,y,w,b):
    m=x.shape[0]
    y1=np.dot(x,w)+b
    cost=y1-y
    cost=(1/(2*m))*np.sum(cost*cost)
    return cost
```

```python
def grad_calc(x,y,w,b):
    m=x.shape[0]
    y1=np.dot(x,w)+b
    cost=y1-y
    w_der=(1/m)*np.sum(x.T@cost)
    b_der=(1/m)*np.sum(cost)
    return w_der,b_der
```

```python
def calculate_grad(x,y,w,b,cost,alfa,itera):
    cost_h=[]
    w_hist=np.array([])
    b_hist=np.array([])
    for item in range(itera):
        w_d,b_d=grad_calc(x,y,w,b)
        w=w-alfa*w_d
        b=b-alfa*b_d
        cost_h.append(cost_function(x,y,w,b))
        w_hist=np.append(w_hist,w)
        b_hist=np.append(b_hist,b)  
    return cost_h,w_hist,b_hist
```
